# Gráfico de Barras

```python
from matplotlib import pyplot as plt
import seaborn as sns
import pandas as pd
```

### Leitura do conjunto de dados

```python
df = pd.read_csv('https://gist.githubusercontent.com/omadson/770934cb0df9e2d2c3e8db031f1139a5/raw/4d181efc035b327d9d66256915951384517c603e/arrecadacao.csv')
df.head()
```

```python
# Simplificando valores
df = df.assign(arrecadacao_milhao = df.arrecadacao/1000000)
```

```python
#Ordena as barras
df = df.sort_values(by='arrecadacao', ascending = False)
```

```python
# Linhas verticais
sns.set_style("whitegrid")
```

## Barras Horizontais

```python
#Mesma cor para as barras (barras horizontais)
plt.figure(figsize=(5,2))
ax = sns.barplot(data = df, x = 'arrecadacao_milhao', y = 'titulo', color = 'tomato')

#Melhora o título e os labels dos eixos
plt.xlabel('arrecadação(milhões de dólares)')
plt.ylabel('Filme')
plt.title('Arrecadação de filmes')

#Adiciona valores na barra
ax.bar_label(
    ax.containers [0],
    fmt = '%.1f',
    label_type = 'center',
    color = 'white'
)

#Remove as bordas e adiciona uma linha vertical na origem
ax.set(frame_on = False)
ax.axvline(0,color = "k",clip_on = False)
plt.show
```

# Gráfico de Histograma

### Leitura do conjunto de dados penguins

```python
df = sns.load_dataset("penguins")
df.head()
```

## Histograma

```python
#linhas verticais
sns.set_style("whitegrid")

#Melora a visualização
ax = sns.histplot(data = df, x = 'body_mass_g')
plt.ylabel('Quantidade')
plt.xlabel('Massa (g)')
plt.title('Distribuição da massa dos pinguins')

#Remove as bordas e adiciona uma linha vertical na origem
ax.set(frame_on = False)
ax.axhline(0,color = "k",clip_on = False)
plt.show
```
