## Estatísticas descritivas dos dados

### Variáveis Quantitativas 

```python
quantitativa_vars = ['total_bill', 'tip', 'size']

media = df[quantitativa_vars].mean()
mediana = df[quantitativa_vars].min()
desvio_padrao = df[quantitativa_vars].std()

print ("\033[1m" +"Média"+ "\033[0m")
print(media)
print(24*"_","\n")

print ("\033[1m" +"Mediana"+ "\033[0m")
print (mediana)
print(24*"_","\n")

print ("\033[1m" +"Desvio Padão"+ "\033[0m")
print (desvio_padrao,"\n")

df.describe ()
```

### Variaveis Qualitativas 

```python
qualitative_vars = ['smoker', 'day', 'time']

for variavel in qualitative_vars:
    valores_unicos = df[variavel].unique()
    for valor in valores_unicos:
        print(f"\033[1m{variavel} - {valor}")
        subamostra = df[df[variavel] == valor]
        display(subamostra.describe())
        print("="*50,"\n")
    
```
